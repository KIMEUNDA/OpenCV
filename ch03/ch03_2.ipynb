```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/girl.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gImg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg.shape)
plt.imshow(gImg, cmap='gray')
```

```python
gPlus = np.zeros((gImg.shape[0], gImg.shape[1]), dtype = np.ubyte)
```

```python
val = 100

for h in range(gImg.shape[0]):
  for w in range(gImg.shape[1]):
    plusVal = gImg[h,w]+ val
    if(plusVal > 255):
      plusVasl = 255

    gPlus[h,w] = plusVal

plt.imshow(gPlus, cmap = 'gray')
```

```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/lena256.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gImg2 = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg.shape)
#plt.imshow(gImg, cmap='gray')
cv2_imshow(gImg2)
```

```python
for h in range(gImg.shape[0]):
  for w in range(gImg.shape[1]):
    plusVal = np.int32(gImg1[h, w]) + np.int32(gImg2[h, w])/2
    if(plusVal > 255):
      plusVasl = 255

    gPlus[h,w] = plusVal

cv2_imshow(gPlus)
```

```python
weight = 0.7


for h in range(gImg.shape[0]):
  for w in range(gImg.shape[1]):
    plusVal = np.int32(gImg1[h, w]) + np.int32(gImg2[h, w])+(1-weight)
    if(plusVal > 255):
      plusVasl = 255

    gPlus[h,w] = plusVal

cv2_imshow(gPlus)
```
