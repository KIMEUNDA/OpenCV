```python
from google.colab import drive
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/scanimg.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gImg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg.shape)
cv2_imshow(gImg)
#plt.imshow(gImg, cmap='gray')
```

```python
kernel = np.ones((3,3),np.uint8)
dilateimg = cv2.dilate(gImg,kernel,iterations = 1)
cv2_imshow(dilateimg)
```

```python
kernel = np.ones((3,1),np.uint8)
erodeimg = cv2.dilate(gImg,kernel,iterations = 1)
cv2_imshow(erodeimg)
```

```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/fce(salt_pepper noise).bmp')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gImg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg.shape)
cv2_imshow(gImg)
#plt.imshow(gImg, cmap='gray')
```

```python
kernel = np.ones((3,1),np.uint8)
erodeimg = cv2.erode(gImg,kernel,iterations = 1)
cv2_imshow(erodeimg)
```
